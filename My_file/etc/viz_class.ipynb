```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
```

```python
warnings.filterwarnings('ignore')
```

```python
plt.rc('font', family='Malgun Gothic')
```

```python
class Viz():
    def __init__(self, df):
        self._df = df

    def __repr__(self):
        return 'DataFrame : {}'.format(df)

    def count(self, x, title):
        plt.figure(figsize=(15,9))
        sns.countplot(data=self._df, x=x)
        plt.title(title, size=20)
        plt.xlabel(x, size=13)
        plt.ylabel('count', size=13)
        plt.tight_layout()
        plt.show()

    def bar(self, x, y, title):
        plt.figure(figsize=(15,9))
        sns.barplot(data=self._df, x=x, y=y)
        plt.title(title, size=20)
        plt.xlabel(x, size=13)
        plt.ylabel(y, size=13)
        plt.tight_layout()
        plt.show()

    def dist(self, x, title, kind='all'):
        plt.figure(figsize=(15, 9))       
        if kind == 'all':
            sns.distplot(x=self._df[x])
        elif kind == 'kde':
            sns.displot(x=self._df[x], kde=True)
        elif kind == 'hist':
            sns.distplot(x=self._df[x], kde=False)
        plt.title(title, size=20)
        plt.xlabel(x, size=13)
        plt.ylabel('count', size=13)
        plt.tight_layout()
        plt.show()

    def kde(self, x, title):
        plt.figure(figsize=(15, 9))
        subset = self._df.loc[self._df['target']==0]
        subset_0 = self._df.loc[self._df['target'] == 0]
        subset_1 = self._df.loc[self._df['target'] == 1]
        subset_2 = self._df.loc[self._df['target'] == 2]
        sns.kdeplot(data=subset_0, x=x)
        sns.kdeplot(data=subset_1, x=x)
        sns.kdeplot(data=subset_2, x=x)
        plt.legend(labels=['0', '1', '2'])
        plt.xlabel(x, size=13)
        plt.ylabel('density', size=13)
        plt.title(title, size=20)
        plt.tight_layout()
        plt.show()

    def box(self, y, title, hue=None):
        plt.figure(figsize=(15, 9))
        if hue == None:
            sns.boxplot(data=self._df, x='target', y=y)
        else:
            sns.boxplot(data=self._df, x='target', y=y, hue=hue)
        plt.xlabel('target', size=13)
        plt.ylabel(y, size=13)
        plt.title(title, size=20)
        plt.tight_layout()
        plt.show()

    def point(self, x, y, title, hue=None):
        plt.figure(figsize=(15, 9))
        if hue == None:
            sns.pointplot(data=self._df, x=x, y=y)
        else:
            sns.pointplot(data=self._df, x=x, y=y, hue=hue)
        plt.xlabel(x, size=13)
        plt.ylabel(y, size=13)
        plt.title(title, size=20)
        plt.tight_layout()
        plt.show()

    def scatter(self, x, y, title, hue=None):
        plt.figure(figsize=(15, 9))
        if hue == None:
            sns.scatterplot(data=self._df, x=x, y=y)
        else:
            sns.scatterplot(data=self._df, x=x, y=y, hue=hue)
        plt.xlabel(x, size=13)
        plt.ylabel(y, size=13)
        plt.title(title, size=20)
        plt.tight_layout()
        plt.show()
```

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
X = iris.data
y = iris.target
names = iris.feature_names
df = pd.DataFrame(data=iris.data, columns=names)
df['target'] = y
df.head()
```

```python
gph = Viz(df)
```

```python
print(gph)
```

```python
gph.count('sepal length (cm)', 'sepal length : count')
```

```python
gph.bar(x='target', y='sepal length (cm)', title='Class_sepal length')
```

```python
gph.dist(x='sepal length (cm)', title='sepal length histogram')
```

```python
gph.dist(x='sepal length (cm)', kind='kde', title='sepal length Histogram')
```

```python
gph.dist(x='sepal length (cm)', kind='hist', title='sepal length Histogram')
```

```python
gph.kde('sepal length (cm)', 'sepal length KDE')
```

```python
gph.kde('sepal width (cm)', 'Sepal Width KDE')
```

```python
gph.box('sepal width (cm)', 'sepal width - class')
```

```python
gph.box('sepal width (cm)', 'sepal width - class')
```

```python
gph.point(x='sepal width (cm)', y='sepal length (cm)', title='Class_sepal length')
```

```python
gph.scatter(x='sepal width (cm)', y='sepal length (cm)', title='Class_sepal length')
```

```python
gph.scatter(x='sepal width (cm)', y='sepal length (cm)', title='Class_sepal length', hue='target')
```
